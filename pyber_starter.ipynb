```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
# city_data_to_load = "data/city_data.csv"
# ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv("./data/city_data.csv")
ride_data_df = pd.read_csv("./data/ride_data.csv")

# Combine the data into a single dataset
combined_data=pd.merge(ride_data_df, city_data_df, on="city")
# Display the data table for preview
combined_data.head()
```

```python
city_counts=combined_data.groupby('city').city.count()
city_fare_avg=combined_data.groupby('city').fare.mean()
driver_counts_per_city=city_data_df['driver_count']
driver_counts_per_city
type_group=combined_data.groupby("type")
type_count=combined_data.groupby("type").count()
type_df=pd.DataFrame({"Total Type": type_count['driver_count']    
})
city_counts
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
x=city_counts
y=city_fare_avg
# Build the scatter plots for each city types
plt.scatter(x,y,marker='o',s=driver_counts_per_city*5, 
            edgecolors="black", cmap="type",
           alpha=0.75)
# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fair ($)")
# Save Figure
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fares_by_type_sum=combined_data.groupby('type').fare.sum()
fares_by_type_sum
# Build Pie Chart
labels=['Rural','Suburban','Urban']
colors=['Gold','SkyBlue','Coral']
explode=(0,0,0.3)

plt.pie(fares_by_type_sum,explode=explode, labels=labels, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Percentage of Total Fares by City Type")
# Save Figure
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_by_city_type=combined_data.groupby('type').city.count()
# Build Pie Chart
labels=['Rural','Suburban','Urban']
colors=['Gold','SkyBlue','Coral']
explode=(0,0,0.3)

plt.pie(rides_by_city_type,explode=explode, labels=labels, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Percentage of Total Rides by City Type")
# Save Figure

rides_by_city_type.head()
```

```python
# Show Figure
plt.show()
```
